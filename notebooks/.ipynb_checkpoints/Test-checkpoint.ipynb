If suppliers of the apples decide to sell apples at $1/lb.  Therefore, they only supply 1000 pounds. 

Price: $1/lb
Therefore 
Quantity supplied = 1000lbs
Quantity demanded = 4000lbs

Therefore, shortage of 3000 apples. 

Price will increase since there's a shortage of apples.
Quantity will also increase because suppliers realize the excess demadn.


Demand Curve:

Law of Demand: inverse relationship betwene price and Quantity demanded.  

Demand schedule - price goes up, quantity demanded goes down.

Reasons curve is downward sloping:
Substitution effect: changes in price motivate consumers to buy relatively cheaper goods.

Income effect: changes in price affec the purchasing power of consumers' income.

Law of diminishing marginal utility: as you contineu to consume a given product, you will eventually get less additional utility/satisfaction from each unit consumed.

If anything other than price changes, tehres a demand shift
